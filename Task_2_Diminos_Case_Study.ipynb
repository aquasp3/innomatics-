```python
import pandas as pd

df = pd.read_csv("diminos_data.csv")
df.head()
```

```python
len(df)
```

```python
df.columns
```

```python
df['order_placed_at'] = pd.to_datetime(df['order_placed_at'])
df['order_delivered_at'] = pd.to_datetime(df['order_delivered_at'])
df.head()
```

```python
df['del_time'] = (
    df['order_delivered_at'] - df['order_placed_at']
).dt.total_seconds() / 60
```

```python
df.head()
```

```python
df['del_time'].describe()
```

```python
df['del_time'].quantile(0.95)
```

```python
import matplotlib.pyplot as plt

plt.boxplot(df['del_time'])
plt.ylabel("Delivery Time (minutes)")
plt.title("Box Plot of Delivery Time")
plt.show()

```
