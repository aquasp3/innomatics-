```python
class Solution(object):
    def defangIPaddr(self, address):
        temp = list(address)
        for i in range(len(temp)):
            if temp[i] == ".":
                temp[i] = "[.]"
        return "".join(temp)

obj = Solution()

m = obj.defangIPaddr("1.1.1.1")
print(m)
```

```python
class Solution(object):
    def findNumbers(self, nums):

        count=0
        for i in nums:
            if len(str(i))%2==0:
                count+=1


        return count
obj = Solution()
m=  obj.findNumbers([12,345,2,6,7896])
print(m)




```

```python
class Solution(object):
    def numIdenticalPairs(self, nums):
        pair=0
        for i in range(len(nums)):

            for j in range(i+1,len(nums)):
                if nums[i]==nums[j]:
                    pair+=1
        return pair
obj = Solution()
m=  obj.numIdenticalPairs([1,2,3,1,1,3])
print(m)
```

```python
class Solution(object):
    def smallerNumbersThanCurrent(self, nums):

        a=[]

        for i in range(len(nums)):
            count=0
            for j in range(len(nums)):
                if nums[j]<nums[i]:
                    count+=1
            a.append(count)
        return a
obj=Solution()
m=obj.smallerNumbersThanCurrent([8,1,2,2,3])
print(m)
```

```python
class Solution(object):
    def subtractProductAndSum(self, n):
        product=1
        sum1=0
        while n>0:
            digit=n%10
            product=product*digit
            sum1=sum1+digit
            n=n//10
        return product-sum1

obj=Solution()
m=obj.subtractProductAndSum(234)
print(m)
```

```python
class Solution(object):
    def xorOperation(self, n, start):
        xor = 0
        for i in range(n):
            xor = xor ^ (start + 2*i)
        return xor
obj=Solution()
m=obj.xorOperation(5,0)
print(m)
```
