```python
class Solution(object):
    def runningSum(self, nums):
        total = 0
        result = []
        for n in nums:
            total += n
            result.append(total)
        return result



nums = list(map(int, input().split()))

obj = Solution()
print(obj.runningSum(nums))
```

```python
class Solution(object):
    def shuffle(self, nums, n):
        result = []
        for i in range(n):
            result.append(nums[i])
            result.append(nums[i+n])
        return result

nums = list(map(int, input().split()))
n = int(input())

obj = Solution()
print(obj.shuffle(nums, n))
```

```python
class Solution(object):
    def kidsWithCandies(self, candies, extraCandies):
        maxCandies = max(candies)
        result = []

        for c in candies:
            if c + extraCandies >= maxCandies:
                result.append(True)
            else:
                result.append(False)

        return result
candies = list(map(int, input().split()))
extra = int(input())

obj = Solution()
print(obj.kidsWithCandies(candies, extra))

```
